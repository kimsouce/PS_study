```python
#2644 촌수 계산
from collections import deque

n = int(input())
a, b = map(int, input().split())
m = int(input())

graph = [[]for i in range(n+1)] # 관계를 저정하는 그래프
info = [0] * (n+1)

for i in range(m):
    p, c =map(int,input().split()) #그래프 저장
    graph[p].append(c)
    graph[c].append(p)
queue = deque()
def bfs(start, end):

    queue.append(start)
    visited = []
    while queue:
        v = queue.popleft()
        visited.append(v)
        if v == end:
            break
        for c in graph[v]:
            if c not in visited:
                info[c] = info[v] + 1
                queue.append(c)
    if info[end] == 0:
        return -1
    else:
        print(info[end])
        
bfs(a,b)
            
```
